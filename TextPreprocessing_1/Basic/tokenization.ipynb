### Regular Expression

```python
import re

text = 'I am going to Delhi!'

tokens = re.findall(r"[|\w+]", text)
tokens
```

```python
import re

text = 'I am going to Delhi!'

tokens = re.findall(r"\w+", text)
tokens
```

```python
sample_text = """
How are you feeling today? 
I hope everything is going well. 
Remember to buy milk, eggs, and bread, 
and don’t forget to call your friend!
"""
```

```python
texts = re.compile('[.!?] ').split(sample_text)
texts
```

### NLTK

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

##### using word tokenize

```python
text = 'I am going to Delhi!'
word_tokenize(text)
```

```python
text = 'I have a Ph.D in A.I'
word_tokenize(text)
```

```python
text = "We're here to help! mail us at nks@gmail.com"
word_tokenize(text)
```

```python
text = 'A 5km ride costs $10.53'
word_tokenize(text)
```

##### using sentence tokenize

```python
sample_text = """
How are you feeling today? 
I hope everything is going well. 
Remember to buy milk, eggs, and bread, 
and don’t forget to call your friend!
"""

sent_tokenize(sample_text)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
text1 = 'I am going to Delhi!'
text2 = 'I have a Ph.D in A.I'
text3 = "We're here to help! mail us at nks@gmail.com"
text4 = 'A 5km ride costs $10.53'
text5 = """
How are you feeling today? 
I hope everything is going well. 
Remember to buy milk, eggs, and bread, 
and don’t forget to call your friend!
"""
```

```python
doc1 = nlp(text1)
doc2 = nlp(text2)
doc3 = nlp(text3)
doc4 = nlp(text4)
doc5 = nlp(text5)
```

```python
for token in doc1:
    print(token)
```

```python
texts = [doc1, doc2, doc3, doc4, doc5]

for i, doc in enumerate(texts, 1):
    print(f"\nTokens in text{i}:")
    for token in doc:
        print(token.text)
```
